**No caso de 18 % 7**

- 18 é o dividendo.
- 7 é o divisor.
- 
O resultado dessa operação é o resto da divisão de 18 por 7. Matematicamente, você pode representar isso como:

18 mod
 

O resultado desse cálculo é 4, porque 18 dividido por 7 é igual a 2 com um resto de 4. 

Portanto, em Python, o valor resultante de 18 % 7 será 4.

### Operador de Comparação


- Estes são usados para comparar dois valores
- Dá um resultado booleano (Verdadeiro/Falso)

Vamos ver algumas comparações numericas abaixo:

```python
2 < 3
```

```python
2 == 3
```

Outras comparações:

```python
"Rahul" < "Rohan"
```

Em Python, quando você compara strings usando o operador de comparação < (menor que), a comparação é feita lexicograficamente, ou seja, com base na ordem alfabética dos caracteres.

No caso da expressão "Rahul" < "Rohan", ela será avaliada como True. Isso ocorre porque, na ordem alfabética, a letra 'a' (de "Rahul") vem antes da letra 'o' (de "Rohan").

Assim, a expressão "Rahul" < "Rohan" é verdadeira em Python.

### Estruturas de dados e tipos de dados


#### Se você estiver aprendendo Python de várias fontes, poderá encontrar os termos estruturas de dados e tipos de dados sendo usados alternadamente.

- **Definição:** Estrutura de dados é um conceito geral da ciência da computação. Sua definição é a seguinte na Wikipédia:
- **A estrutura de dados** é um formato de organização, gerenciamento e armazenamento de dados que permite acesso e modificação eficientes. Mais precisamente, uma estrutura de dados é uma coleção de valores de dados, suas relações e as funções ou operações que podem ser aplicadas aos dados.
- Considerando que tipo de dados é um conceito específico de uma linguagem de programação. De certa forma, é uma implementação concreta de uma estrutura de dados em uma determinada linguagem de programação (Python ou qualquer outra linguagem).
- A definição do que constitui um "tipo" varia entre as linguagens de programação. Falando em Python, existem tipos de dados básicos como int, float, string, etc. Você pode usar os tipos internos como lista, conjunto, etc., que abordaremos nesta sessão.
- Chamar esses tipos de dados como estruturas de dados não será errado porque não há diferença significativa entre os dois em Python.

### Fatiamento de cadeia de caracteres


- Também podemos chamar uma variedade de caracteres da cadeia de caracteres usando fatiamento de cadeia de caracteres.
- Especifique os índices inicial e final separados por dois pontos para retornar uma parte da cadeia de caracteres. Observe que o caractere do índice final não está incluído.
- Suponha que queremos imprimir World a partir da cadeia de caracteres "Hello World". 


Podemos fazê-lo da seguinte forma:

```python
a = "Hello World!"
print(a[-4])
```

### Indexação Negativa


- Se tivermos uma cadeia de caracteres longa e quisermos identificar um item no final, também podemos contar para trás a partir do final da cadeia de caracteres, **começando com o número de índice -1**.
- Imprimindo 'r' a partir da cadeia de caracteres observado acima.

Agora vamos obter os caracteres da posição -5 para a posição -1, iniciando a contagem a partir do final da cadeia de caracteres:

```python
print(a[-5:-2])
```
