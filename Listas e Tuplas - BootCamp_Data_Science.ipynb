### Lista em Python


Como o nome sugere, List é uma sequência ordenada de dados. Na vida real, se você pudesse fazer uma lista de coisas que vêm à sua mente (ou evento para qualquer propósito específico), poderia ser algo assim:

- Escova
- Leuven
- 48851964400
- 3.14
- Mamãe


**Bem, esta é a minha lista.** Você pode fazer sua lista e incluir o que quiser nela. Então, na minha lista, eu incluí o que eu faço de manhã cedo, minha cidade, meu número de celular, o valor de pi a dois dígitos e a mãe. Ele tem diferentes tipos de dados – strings, float e inteiro.

- **Esse é o tipo de flexibilidade que o Python List oferece.** Ele pode conter diferentes tipos de tipos de dados. Declarar uma Lista é bastante simples. **Use colchetes ([]) e separe os itens por uma vírgula.** 
 

Exemplo:


A = ["Pincel", "Lovaina", 48851964400, 3.14, "Mãe"]

- As listas são mutáveis. Digamos que se você quiser alterar algum item em uma lista, você pode fazer isso. Por exemplo, se eu não gosto de 'Brush' e quero substituir isso por 'Morning Walk'.


A = ["Morning Walk", "Leuven", 48851964400, 3.14, "Mom"]

#### Algumas características essenciais das listas Python são:

* Coleta de valores
* Pode ser de qualquer tipo de dados
* Pode ser uma combinação de diferentes tipos

### Métodos de lista

- .append()


No Python, você pode adicionar valores ao final de uma lista usando o método .append().

```python
orders = ['daisies', 'periwinkle']
orders
```

```python
orders.append('tulips')
# Adicioando ao fim da sua lista
```

```python
print(orders)
```

- count () 

O método de lista .count() Python pesquisa em uma lista qualquer termo de pesquisa que ela receba como argumento e, em seguida, retorna o número de entradas correspondentes.

```python
backpack = ['pencil', 'pen', 'textbook', 'pen', 'highlighter', 'pen']
```

```python
numPen = backpack.count('pen')
# Contagem do número de vezes que a várivel 'pen' ocorre na lista
```

```python
print(numPen)
```

- len()

A função len() do Python pode determinar o número de itens encontrados na lista que aceita como argumento.

```python
backpack = ['pencil', 'pen', 'notebook', 'textbook', 'pen', 'highlighter', 'pen']
print(len(backpack))
```

- .sort()
  
O método de lista .sort() Python classificará o conteúdo de qualquer lista em que ele for chamado, ou seja, as listas numéricas serão ordenadas em ordem crescente e as listas de Strings serão ordenadas em ordem alfabética. 

- Ele modifica a lista original e não tem valor de retorno.

```python
exampleList = [4, 2, 1, 3]
```

```python
exampleList.sort()
```

```python
print(exampleList)
```

### Tuplas em Python

- Tupla também é uma sequência ordenada de itens como uma lista. Tupla também contém vários tipos de dados.
- A única diferença entre Tuple & List é que Tuple é imutável; Uma vez criado, ele não pode ser alterado.
- Criar uma tupla é tão simples quanto colocar diferentes valores separados por vírgulas entre colchetes redondos.

Exemplo: A = ('Pincel', 'Lovaina', 48851964400, 3.14, 'Mãe')
